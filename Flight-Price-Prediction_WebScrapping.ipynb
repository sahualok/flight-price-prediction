# NEW DELHI TO MUMBAI

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
```

```python
# Opening page of yatra.com for flights from new delhi to mumbai
url = "https://flight.yatra.com/air-search-ui/dom2/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BOM&destinationCountry=IN&flight_depart_date=21%2F01%2F2022&ADT=1&CHD=0&INF=0&class=Economy&source=fresco-home&unqvaldesktop=31443714689"
driver.get(url)
```

```python
# creating empty lists for scraping data
airline_name=[]
source=[]
destination=[]
departure_time=[]
arrival_time=[]
travel_duration=[]
total_stops=[]
price=[]


#scraping the airlines name
name=driver.find_elements_by_xpath("//span[@class='i-b text ellipsis']")
for i in name:
    if i.text is None :
        airline_name.append("--") 
    else:
        airline_name.append(i.text)
        
#scraping the source
source_name=driver.find_elements_by_xpath("//p[@title='New Delhi']")
for i in source_name:
    if i.text is None :
        source.append("--") 
    else:
        source.append(i.text)
        
#scraping the destination
dest_name=driver.find_elements_by_xpath("//p[@title='Mumbai']")
for i in dest_name:
    if i.text is None :
        destination.append("--") 
    else:
        destination.append(i.text)
        
#scraping the departure time
time=driver.find_elements_by_xpath("//div[@class='i-b pr']")
for i in time:
    if i.text is None :
        departure_time.append("--") 
    else:
        departure_time.append(i.text)
        
#scraping the arrival time
arrtime=driver.find_elements_by_xpath("//p[@class='bold fs-15 mb-2 pr time']")
for i in arrtime:
    if i.text is None :
        arrival_time.append("--") 
    else:
        arrival_time.append(i.text.replace("\n+ 1 day",""))
        
#scraping the duration
duration=driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']")
for i in duration:
    if i.text is None :
        travel_duration.append("--") 
    else:
        travel_duration.append(i.text)
        
#scraping total stops
stops=driver.find_elements_by_xpath(" //div[@class=' font-lightgrey fs-10 tipsy i-b fs-10']")
for i in stops:
    if i.text is None :
        total_stops.append("--") 
    else:
        total_stops.append(i.text)
        
#scraping the price
prc=driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']")
for i in prc:
    if i.text is None :
        price.append("--") 
    else:
        price.append(i.text)

# creating the dataframe from the scraped data
df=pd.DataFrame({"Airline Name":airline_name,"Source":source,"Destination":destination,"Departure Time":departure_time,"Arrival Time":arrival_time,"Duration":travel_duration,"Total Stops":total_stops,"Price":price})
df      
```

```python
df.to_csv("FlightPrice1", index=False)
```

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
```

```python
# Opening page of makemytrip.com for flights from new delhi to bangalore
url = "https://www.makemytrip.com/flight/search?itinerary=DEL-BLR-22/01/2022&tripType=O&paxType=A-1_C-0_I-0&intl=false&cabinClass=E&ccde=IN&lang=eng"
driver.get(url)
```

```python
# creating empty lists for scraping data
airline_name=[]
source=[]
destination=[]
departure_time=[]
arrival_time=[]
travel_duration=[]
total_stops=[]
price=[]


#scraping the airlines name
name=driver.find_elements_by_xpath("//span[@class='boldFont blackText airlineName']")
for i in name:
    if i.text is None :
        airline_name.append("--") 
    else:
        airline_name.append(i.text)
        
#scraping the source
source_name=driver.find_elements_by_xpath("//div [@class='timingOptionOuter']/label/div/div/div[1]/div[1]/p[2]")
for i in source_name:
    if i.text is None :
        source.append("--") 
    else:
        source.append(i.text)
        
#scraping the destination
dest_name=driver.find_elements_by_xpath("//div[@class='flightTimeSection flexOne timeInfoRight']/div[1]/p[2]")
for i in dest_name:
    if i.text is None :
        destination.append("--") 
    else:
        destination.append(i.text)
        
#scraping the departure time
time=driver.find_elements_by_xpath("//div [@class='timingOptionOuter']/label/div/div/div[1]/div[1]/p[1]")
for i in time:
    if i.text is None :
        departure_time.append("--") 
    else:
        departure_time.append(i.text)
        
#scraping the arrival time
arrtime=driver.find_elements_by_xpath("//div[@class='flightTimeSection flexOne timeInfoRight']/div[1]/p[1]/span")
for i in arrtime:
    if i.text is None :
        arrival_time.append("--") 
    else:
        arrival_time.append(i.text)
        
#scraping the duration
duration=driver.find_elements_by_xpath("//div[@class='stop-info flexOne']/p")
for i in duration:
    if i.text is None :
        travel_duration.append("--") 
    else:
        travel_duration.append(i.text)
        
#scraping total stops
stops=driver.find_elements_by_xpath("//p[@class='flightsLayoverInfo']")
for i in stops:
    if i.text is None :
        total_stops.append("--") 
    else:
        total_stops.append(i.text)
        
#scraping the price
prc=driver.find_elements_by_xpath("//p[@class='blackText fontSize18 blackFont white-space-no-wrap']")
for i in prc:
    if i.text is None :
        price.append("--") 
    else:
        price.append(i.text.replace("₹ ",""))

# creating the dataframe from the scraped data
df=pd.DataFrame({"Airline Name":airline_name,"Source":source,"Destination":destination,"Departure Time":departure_time,"Arrival Time":arrival_time,"Duration":travel_duration,"Total Stops":total_stops,"Price":price})
df      
```

```python
df.to_csv("FlightPrice6", index=False)
```
